## Cost Function

#### What is cost function?
A cost function is a measure of how well a machine learning model performs by quantifying the difference between predicted and actual outputs.

#### Mathematical Expression:

The equation for cost with one variable is:

ùêΩ(ùë§,ùëè)=1/2ùëö*(i=0,m-1 ‚àë [ùëìùë§,ùëè(ùë•(ùëñ))‚àíùë¶(ùëñ)]^2)

ùëìùë§,ùëè(ùë•(ùëñ))=ùë§ùë•(ùëñ)+ùëè

where,
* ùëìùë§,ùëè(ùë•(ùëñ)) is the prediction for example ùëñ using parameters ùë§,ùëè
* (ùëìùë§,ùëè(ùë•(ùëñ))‚àíùë¶(ùëñ))^2 is the squared difference between the target value and the prediction. 
* These differences are summed over all the ùëö examples and divided by 2m to produce the cost, ùêΩ(ùë§,ùëè)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y = np.array([250, 300, 480, 430, 630, 730])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
def cost_function(x,y,w,b):
    
    m = x.shape[0] #size of array
    cost_sum = 0
    
    for i in range(m):
        f_wb = w*x[i]+b
        cost = (f_wb-y[i])**2
        cost_sum = cost_sum+cost
        
    total_cost = (1/(2*m))*cost_sum
    
    return total_cost
```

```python
w = 200
b = 25

cost = cost_function(x,y,w,b)

print(f"Cost Function for w: {w} and b: {b} is {cost}")
```

### Visualizing the Model

```python
predicted = np.zeros(6)
for i in range(6):
        predicted[i] = w*x[i]+b
```

```python
plt.scatter(x,predicted)
plt.plot(x,predicted)
plt.title("Current Model")
plt.xlabel('Actual values of x')
plt.ylabel('Predicted values of y')
plt.show()
```
